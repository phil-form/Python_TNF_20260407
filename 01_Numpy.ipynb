# Concatener

```python
arr_1 = np.array([1, 2, 3])
arr_2 = np.array([4, 5, 6])
arr_3 = np.array([7, 8, 9])

arr_concat = np.concatenate(
    [arr_1, arr_2, arr_3],
)

arr_concat
```

```python
arr_concat = np.concatenate(
    [
        arr_1.reshape(1, -1),
        arr_2.reshape(1, -1),
        arr_3.reshape(1, -1)
    ],
)

arr_concat
```

```python
arr_concat = np.concatenate(
    [
        arr_1.reshape(1, -1),
        arr_2.reshape(1, -1),
        arr_3.reshape(1, -1)
    ],
    axis=0
)

arr_concat
```

```python
arr_concat = np.concatenate(
    [
        arr_1.reshape(-1, 1),
        arr_2.reshape(-1, 1),
        arr_3.reshape(-1, 1)
    ],
    axis=1
)

arr_concat
```

```python
arr_concat = np.concatenate(
    [
        arr_1.reshape(-1, 1),
        arr_1.reshape(-1, 1),
        arr_2.reshape(-1, 1),
        arr_2.reshape(-1, 1),
        arr_3.reshape(-1, 1),
        arr_3.reshape(-1, 1),
    ],
    axis=1
)

arr_concat
```
