# Jointure 

```python
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": [
        "Alice",
        "Bob",
        "Charlie"
    ]
})

df2 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "score": [
        25,
        12,
        32,
        13
    ]
})

result = pd.merge(df1, df2, on="id", how="inner")
result
```

```python
pd.merge(df1, df2, on="id", how="outer")
```

```python
pd.merge(df1, df2, how="cross")
```

```python
pd.merge(df1, df2, on="id", how="left")
```

```python
pd.merge(df1, df2, on="id", how="right")
```

# Column Conversion

```python
df3 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "name": [
        "Alice",
        "Bob",
        "Charlie",
        "David",
    ],
    "birthdate": [
        "19901211",
        "19900302",
        "19901102",
        "19900304",
    ]
})

df3
```

```python
df3['date-time'] = pd.to_datetime(df3['birthdate'])
df3
```

```python
df3['timestamp'] = pd.to_datetime(df3['birthdate']).astype(int)
df3
```

```python
df4 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "name": [
        "Alice",
        "Bob",
        "Charlie",
        "David",
    ],
    "birthdate": [
        "11/12/1990 12:00:00.000",
        "02/03/1990 12:00:00.000",
        "02/11/1990 12:00:00.000",
        "04/03/1990 12:00:00.000",
    ]
})

df4
```

```python
df4['datetime'] = pd.to_datetime(df4['birthdate'], format="%d/%m/%Y %H:%M:%S.%f")
df4
```
