```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
homeprices = pd.read_csv('Datasets/homeprices practice.csv')
homeprices.head()
```

### Visualization 📊


```python
plt.xlabel('Area (in m2)')
plt.ylabel('Price (in $)')

plt.scatter(x=homeprices['area'],
            y=homeprices['price'], color='red', marker='+')
plt.show()
```

```python
# Separate Target and Features
X = homeprices[['area']]
y = homeprices['price']
```

### Fitting the Model 🎡

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()  # Create a model object
model.fit(X, y)  # Fit the model
```

```python
model.score(X, y)  # R^2 value, coefficient of determination
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[2600]])
```

```python
model.predict([[0]])
```

```python
X
```

```python
model.predict(X)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculer les predictions
y_pred = model.predict(X)


display(f"Coef determination : {r2_score(y, y_pred)}")
display(f"Erreur absolue Moyenne : {mean_absolute_error(y, y_pred)}")
f"Erreur quadratique Moyenne : {mean_squared_error(y, y_pred)}"
```

```python
mean_squared_error(y, y_pred)**0.5  # RMSE
```

```python
def predict_price(area):
    return f"The price of a {area} squared meter is : {model.predict([[area]])[0]} $"
```

```python
predict_price(3000)
```
